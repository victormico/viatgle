```python
import json
import os
```

```python
cwd = os.getcwd()
parent = os.path.dirname(cwd)
src_path = os.path.join(parent, 'src')
game_info_file = os.path.join(src_path, 'games', 'game.json')
```

```python
#read the data from the file
with open("comarques_limitrofes.json") as f:
    comarques_adjacents = json.load(f)
```

```python
comarques_list = list(comarques_adjacents.keys())
```

```python
#read the data from the file
with open("all_shortest_paths.json") as f:
    all_shortest_paths = json.load(f)

game = {}

game["start"] = "alt_emporda"
game["end"] = "vall_d_albaida"
game["shortests_paths"] = all_shortest_paths[game["start"]][game["end"]]



for path in game["shortests_paths"]:
    path.remove(game["start"])
    path.remove(game["end"])
    print(path)

# save game to a dict file
with open(game_info_file, "w") as f:
    json.dump(game, f, indent=4)
```

```python
# Get data from game dictionary
start = game["start"]
end = game["end"] 
shortests_paths = game["shortests_paths"]

# Inicialització
guesses_icons = {
    "optimal": "✅",
    "good": "🟩",
    "pretty_good": "🟧",
    "bad": "🟥"
}


remaining_guesses = len(shortests_paths[0]) + 3
guesses = []
guesses_status = []
progress = {idx: path.copy() for idx, path in enumerate(shortests_paths)}  # Mantenir els camins
game_running = True  # Bandera per mantenir el joc en marxa

print(f"I want to go from {start} to {end}")
while game_running and remaining_guesses > 0:
    guess_icon = guesses_icons['bad']
    guess_status = f"{guess_icon} Guess is not great. This comarca is a significant detour."
    print(f"Number of guesses remaining: {remaining_guesses}")
    
    # Demanar una comarca
    while game_running:
        guess = input("Please enter a comarca id (or type 'exit' to quit): ").strip().lower()
        if guess == "exit":
            print("You have chosen to exit the game. Goodbye!")
            game_running = False
            break
        elif guess not in comarques_list:
            print("Guess is not a comarca id")
        elif guess in guesses:
            print(f"You have already guessed this comarca: {guess}")
        else:
            guesses.append(guess)
            print(f"You guessed: {guess}")  # Debugging: Mostrar la comarca introduïda
            break
    
    if not game_running:
        break  # Sortir del loop principal també si l'usuari ha escrit 'exit'
    
    # Verificar el nivell del guess
    optimal = False
    good = False
    pretty_good = False

    for path in progress.values():
        if path:
            if guess == path[0]:  # ✅ Optimal guess
                optimal = True
                path.pop(0)  # Treure del camí perquè ja s'ha encertat
            elif guess in path:  # 🟩 Good guess
                good = True
                # get index of the guess in path
                idx = path.index(guess)                
                path.pop(idx)  # Treure del camí perquè s'ha encertat, però no és el següent
            elif any(guess in comarques_adjacents[c] for c in path):  # 🟧 Pretty good guess
                pretty_good = True
    
    # Assignar el millor estat
    if optimal:
        guess_icon = guesses_icons['optimal']
        guess_status = f"{guess_icon} Guess is optimal! You're one step closer."
        
    elif good:
        guess_icon = guesses_icons['good']
        guess_status = f"{guess_icon} Guess is good. This guess is on the shortest path, but it's not the next comarca to go through."
    elif pretty_good:
        guess_icon = guesses_icons['pretty_good']
        guess_status = f"{guess_icon} Guess is pretty good. This comarca is close, but doesn't get you closer."
    
    print(guess_status)
    guesses_status.append(guess_icon)
    print(guesses_status)

    # Comprovar si algun camí ha estat completat
    if any(len(path) == 0 for path in progress.values()):
        print("You have won!")
        game_running = False
        break

    # Reduir intents restants
    remaining_guesses -= 1

# Missatge final si el joc ha acabat
if not any(len(path) == 0 for path in progress.values()):
    print("Game over! You have run out of guesses.")
```
