```python
import json
```

```python
#read the data from the file
with open("comarques_limitrofes.json") as f:
    comarques_adjacents = json.load(f)
```

```python
comarques_list = list(comarques_adjacents.keys())
```

```python
#read the data from the file
with open("all_shortest_paths.json") as f:
    all_shortest_paths = json.load(f)

start = "alacanti"
end = "alcalaten"
shortests_paths = all_shortest_paths[start][end]
for path in shortests_paths:
    path.remove(start)
    path.remove(end)
    print(path)
# shortests_paths
```

```python
# InicialitzaciÃ³
remaining_guesses = len(shortests_paths[0]) + 3
guesses = []
progress = {idx: path.copy() for idx, path in enumerate(shortests_paths)}  # Mantenir els camins
game_running = True  # Bandera per mantenir el joc en marxa

print(f"I want to go from {start} to {end}")
while game_running and remaining_guesses > 0:
    guess_status = "ðŸŸ¥ Guess is not great. This comarca is a significant detour."
    print(f"Number of guesses remaining: {remaining_guesses}")
    
    # Demanar una comarca
    while game_running:
        guess = input("Please enter a comarca id (or type 'exit' to quit): ").strip().lower()
        if guess == "exit":
            print("You have chosen to exit the game. Goodbye!")
            game_running = False
            break
        elif guess not in comarques_list:
            print("Guess is not a comarca id")
        elif guess in guesses:
            print(f"You have already guessed this comarca: {guess}")
        else:
            guesses.append(guess)
            print(f"You guessed: {guess}")  # Debugging: Mostrar la comarca introduÃ¯da
            break
    
    if not game_running:
        break  # Sortir del loop principal tambÃ© si l'usuari ha escrit 'exit'
    
    # Verificar el nivell del guess
    optimal = False
    good = False
    pretty_good = False

    for path in progress.values():
        if path:
            if guess == path[0]:  # âœ… Optimal guess
                optimal = True
                path.pop(0)  # Treure del camÃ­ perquÃ¨ ja s'ha encertat
            elif guess in path:  # ðŸŸ© Good guess
                good = True
            elif any(guess in comarques_adjacents[c] for c in path):  # ðŸŸ§ Pretty good guess
                pretty_good = True
    
    # Assignar el millor estat
    if optimal:
        guess_status = "âœ… Guess is optimal! You're one step closer."
    elif good:
        guess_status = "ðŸŸ© Guess is good. This guess is on the shortest path, but it's not the next comarca to go through."
    elif pretty_good:
        guess_status = "ðŸŸ§ Guess is pretty good. This comarca is close, but doesn't get you closer."
    
    print(guess_status)
    
    # Comprovar si algun camÃ­ ha estat completat
    if any(len(path) == 0 for path in progress.values()):
        print("You have won!")
        game_running = False
        break

    # Reduir intents restants
    remaining_guesses -= 1

# Missatge final si el joc ha acabat
if not any(len(path) == 0 for path in progress.values()):
    print("Game over! You have run out of guesses.")
```
