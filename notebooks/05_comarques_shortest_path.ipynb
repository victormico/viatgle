```python
import matplotlib.pyplot as plt
import networkx as nx
import json
import os
```

```python
cwd = os.getcwd()
parent = os.path.dirname(cwd)
src_path = os.path.join(parent, 'src')
all_shortest_paths_path = os.path.join(src_path, 'all_shortest_paths.json')
```

```python
#read the data from the file
with open("comarques_limitrofes.json") as f:
    comarques_data = json.load(f)
```

```python
G = nx.Graph()

# Add edges from adjacency data
for comarca, neighbors in comarques_data.items():
    for neighbor in neighbors:
        G.add_edge(comarca, neighbor)  # By default, weight is 1 for unweighted graphs
```

```python
G
plt.figure(figsize=(8,16))
nx.draw(
    G,
    with_labels=True,
    node_size=50,
    font_size=10
    )
plt.show()
```

```python
all_shortest_paths = dict(nx.all_pairs_all_shortest_paths(G))
```

```python
with open(all_shortest_paths_path, 'w') as f: 
    json.dump(all_shortest_paths, f, indent=4)
```
